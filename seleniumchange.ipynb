```python
#获取页面内容
from selenium import webdriver
browser = webdriver.Chrome()
browser.maximize_window()
lists1,lists2,lists3 = [],[],[]

for i in range(1,11): #i循环修改index=i的值
    url='https://www.weimao.com/search/category/12'+ '?index='+str(i) 
    for j in range(0,10): #j循环遍历整页内容
        browser.get(url)
        input1 = browser.find_elements_by_class_name('search-left-content')[j]
        lists1.append(input1.text)
        input2 = browser.find_elements_by_xpath(".//*[@class='status']")[j]
        lists2.append(input2.text)
        
        
for a,b in map(list,zip(lists1,lists2)):
    lists3.append(a)
    lists3.append(b)
        
browser.close()
```

```python
#打印lists3
print(lists3)
```

```python
i,j,k=0,0,0
afterlist=[]

for num in lists3:
    if (i%2)==0: #公司信息处理
        inflist=num.split()
        print(inflist)
        
        if (k%2)==0:
            
            for index in inflist:
                if(j%2)==0:
                    afterlist.append(index)
                j+=1
                
        else:
            
            for index in inflist:
                if(j%2)!=0:
                    afterlist.append(index)
                j+=1
        k+=1           
    i+=1
print(afterlist)
    
```

```python
print(len(afterlist))
```

```python
#division
div6list=[]
div6listall=[]

for index in range(0,len(afterlist),6):
    div6list=afterlist[index:index+6]
    div6listall.append(div6list)
    
    
    
print(div6listall)
```

```python
print(div6listall[0][0])
print(len(div6listall))
```
